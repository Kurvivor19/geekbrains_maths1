```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
# Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)
from math import sqrt
def veclen(v):
    lsq = 0
    for c in v:
        lsq += c * c
    return sqrt(lsq)
```

```python
# уравнения в параметрической форме
a = 5
b = 3
t = np.linspace(0, 2*np.pi, 500)
x = a * np.cos(t)
y = b * np.sin(t)

plt.plot(x, y)
r = 4
x1 = r * np.cos(t)
y1 = r * np.sin(t)

plt.plot(x1, y1, color='red')

th = np.linspace(-1, 1, 100)
x2 = a * np.cosh(th)
y2 = b * np.sinh(th)
plt.plot(x2, y2, color='green')
plt.plot(-x2, y2, color='green')
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = Y = np.linspace(-10, 10, 50)
XV, YV = np.meshgrid(X, Y)

a = 1
b = -2
d = 5
ax = plt.figure().add_subplot(111, projection='3d')
ax.plot_wireframe(XV, YV, a * XV + b * YV + d, rcount=4, ccount=4)
a = 2
b = -9
d = 1
ax.plot_wireframe(XV, YV, a * XV + b * YV + d, color='yellow', rcount=4, ccount=4)
```

```python
phi = rho = np.linspace(0, 2*np.pi, 50)
phiV, rhoV = np.meshgrid(phi, rho)

a = 5
b = 3
c = 4

XV2 = a * np.cos(phiV) * np.cos(rhoV)
YV2 = b * np.sin(phiV) * np.cos(rhoV)
ZV2 = c * np.sin(rhoV)

ax2 = plt.figure().add_subplot(111, projection='3d')
ax2.plot_wireframe(XV2, YV2, ZV2, color='green', rcount=5, ccount=5)
```
