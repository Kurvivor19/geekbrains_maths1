```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
# Нарисуйте график функции:
# y(x) = k∙cos(x – a) + b
x = np.linspace(-5 * np.pi, 5*np.pi, 250)
k = 1
a = np.pi/2
b = 1.5
plt.plot(x, k * np.cos(x - a) + b, color='yellow')
k = 2
a = - np.pi
b = -3
plt.plot(x, k * np.cos(x - a) + b, color='green')
```

```python
def polar_to_dec(Phi, R):
    from math import cos, sin
    for p,r in zip(Phi, R):
        yield (r * cos(p), r * sin(p))
        

phi = np.linspace(0, 2 * np.pi, 360)
r = np.full(np.shape(phi), 1.5)


x, y = list(zip(*polar_to_dec(phi, r)))
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(x, y)
```

```python
r_0 = 4
slope = np.pi / 6
r2 = r_0 / np.cos(phi - slope)
x2, y2 = list(zip(*polar_to_dec(phi, r2)))
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(x2, y2, color='purple')
```

```python
# Задание (в программе)
# Решите систему уравнений, неравенств
x = np.linspace(-2, 5, 250)
y1 = x ** 2 - 1
y2 = 1 - (1 - np.exp(x)) / x
plt.plot(x, y1, color='green')
plt.plot(x, y2, color='red')
```

```python
from scipy.optimize import fsolve
xfunc = lambda t : 2 - (1 - np.exp(t)) / t - t ** 2
X1 = fsolve(xfunc, -1.5)
Y1 = X1 ** 2 - 1
X2 = fsolve(xfunc, 2.5)
Y2 = X2 ** 2 - 1
X3 = fsolve(xfunc, 4)
Y3 = X3 ** 2 - 1

print("Roots are: (%f, %f), (%f, %f), (%f, %f)" % (X1, Y1, X2, Y2, X3, Y3))
```

Неравенство будет верным для отрезков параболы, которые находятся под красным графиком, т.е. между 1 и вторым корнями, и от 3 корня и до бесконечности
